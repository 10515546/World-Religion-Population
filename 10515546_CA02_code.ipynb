```python
import altair as alt
alt.renderers.enable('notebook')
interval = alt.selection_interval()

points = alt.Chart(wp_reg).mark_circle().encode(
  x='total muslims:O',
  y='year:O', 
  color=alt.condition(interval, 'region', alt.value('lightgray'))
).properties(
    title='Islam Religion scatter plot',
  selection=interval
)

histogram = alt.Chart(wp_reg).mark_bar().encode(
  x='count()',
  y='region',
  color='region'
).transform_filter(interval)  


b = points & histogram
b
b.save('Islam Religion Region Wise.html')
b
```

```python
source = wp_reg

brush = alt.selection(type='interval', resolve='global')

base = alt.Chart(source).mark_circle().encode(
    y='year:O',
    color=alt.condition(brush, 'region', alt.ColorValue('gray'))
).add_selection(
    brush
).properties(
    title='Christianity vs Islam ',
    width=300,
    height=300
)

base1 = alt.Chart(source).mark_circle().encode(
    y='year:O',
    color=alt.condition(brush, 'region', alt.ColorValue('gray'))
).add_selection(
    brush
).properties(
    title='Hindu vs Budhism ',
    width=300,
    height=300
)


a= base.encode(x='total christians:O') | base.encode(x='total muslims:O') |base1.encode(x='total hindus:O')|base1.encode(x='total buddhists:O')
a.save('Top four Religion Comparison.html')
a
```
